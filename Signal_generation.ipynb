# ***Генерация сигналов.***

Свойства сигнала:


1.   Ограничен по времени. Сигнал существует в течении интервала от c1 до d1.
2.   Амплитуда уменьшается при приближении к моменту t1 и увеличивается при отдалении от него
3. Функция нечетная





Построим функцию

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter values
a1 = 1
b1 = 0.1
t1 = 2
w1 = 3
c1 = 2
d1 = 10

def signal(t):
  """Calculates the value of the signal at time t"""

  # Hi-function
  if (c1 <= t) and (d1 >= t):
    ind = 1
  else:
    ind = 0

  return (a1 + b1 * abs(t - t1)) * np.cos(2 * np.pi * w1 * t) * ind


def plotting_signal(t_start, t_end, sampling_rate, signal):
  t = np.linspace(t_start, t_end, int((t_end - t_start) * sampling_rate), endpoint=False)
  x1 = np.array([signal(ti) for ti in t])
  plt.figure(figsize=(6, 3))
  plt.plot(t, x1)
  plt.xlabel("Время (t)")
  plt.ylabel("Амплитуда сигнала x1(t)")
  plt.title("Смоделированный сигнал x1(t)")
  plt.grid(True)
  plt.axvline(x=c1, color='r', linestyle='--', label=f"c1 = {c1}")
  plt.axvline(x=d1, color='g', linestyle='--', label=f"d1 = {d1}")
  plt.axvline(x=t1, color='m', linestyle=':', label=f"t1 = {t1}")
  plt.legend()
  plt.xlim(t_start, t_end)
  plt.show()

```

Сделаем преобразование фурье

```python
t_start = 5
t_end = 15
sampling_rate = 100

def fourier_analysis(t_start, t_end, sampling_rate, signal):
  N = int((t_end - t_start) * sampling_rate)
  t = np.linspace(t_start, t_end, N, endpoint=False)
  x1 = np.array([signal(ti) for ti in t])

  xf = np.fft.fft(x1) # Преобразование Фурье
  freq = np.fft.fftfreq(x1.size, d=1/sampling_rate)
  xf = np.fft.fftshift(xf)
  freq = np.fft.fftshift(freq)

  yf = np.abs(xf)

  plt.figure(figsize=(6, 3))
  plt.plot(freq, yf)
  plt.xlabel("Частота (Гц)")
  plt.ylabel("Амплитуда")
  plt.title("Спектр Фурье сигнала x1(t)")
  plt.grid(True)
  plt.xlim(-10, 10)
  plt.show()

fourier_analysis(t_start, t_end, sampling_rate, signal)
```

Давайте поизменяем параметры *t_start, t_end, sampling_rate* и сравним вид функции

```python
t_start_1 = -1
t_end_1 = 11
sampling_rate_1 = 10

print("Вариант 1")
plotting_signal(t_start_1, t_end_1, sampling_rate_1, signal)
fourier_analysis(t_start_1, t_end_1, sampling_rate_1, signal)
```
